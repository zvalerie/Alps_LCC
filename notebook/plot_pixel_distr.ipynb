```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data
categories = [
    "Scree",
    "Bedrock",
    "Glacier",
    "Forest",
    "Scree with grass",
    "Water",
    "Bedrock with grass",
    "Large blocks",
    "Large blocks with grass",
    "Background",
]

values = [
2685660429.00,
2646292039.00,
2346320913.00,
1167852467.00,
130810865.00,
77478467.00,
60248636.00,
23840528.00,
2432527.00,
40463129.00,
]


```

```python
# Grouping and color mapping
categories_group = ["Head", "Head", "Head", "Body", "Body", "Body", "Tail", "Tail",'Background',]
colors =    [ 'tab:blue'] *4 +  ['tab:green'] * 3 + ['tab:orange'] *2 + ["grey"]
```

```python

# Create a logarithmic scale y-axis
#values_log = np.log10(values)
plt.rcParams['font.family'] = ['Times New Roman']
# Increase font size for various elements
plt.rc('axes', titlesize=18)      # Title font size
plt.rc('axes', labelsize=14)      # Axis labels font size
plt.rc('xtick', labelsize=14)     # X-axis tick labels font size
plt.rc('ytick', labelsize=12)     # Y-axis tick labels font size
plt.rc('legend', fontsize=10)     # Legend font size


# Create the bar plot
plt.figure(figsize=(6, 5))
plt.bar(categories, values, color=colors,)
#plt.xlabel("Log Scale (base 10)")
plt.ylabel("Pixel count on log scale")
plt.yscale("log")
plt.title("Distribution of pixel per class", fontsize=14)

# Customize the legend
legend_labels = {"Head": "tab:blue", "Body": "tab:green",'Tail':'tab:orange', 'Background':'grey'}
handles = [plt.Rectangle((0, 0), 1, 1, color=color, label=label) for label, color in legend_labels.items()]
plt.legend(handles=handles,)
plt.xticks(rotation=90, ha="right")
plt.ylim(5E+5, 1E+10)

# Save the plot as a vector image (e.g., PDF)
plt.tight_layout()
# Show the plot

plt.savefig("pixel_distribution.pdf", format="pdf")
plt.show()
```
