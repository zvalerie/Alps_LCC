```python
import matplotlib.pyplot as plt

# Sample data
# Extract x and y values from tuples
x1, y1 = zip(*val_coord)
x2, y2 = zip(*test_coord)
x3, y3 = zip(*train_coord)

s = 0.1

# Create a scatter plot
plt.figure(figsize=(20, 10)) 

plt.scatter(x3, y3, color='blue', label='train', s=s, marker='.')
plt.scatter(x2, y2, color='green', label='test', s=s, marker='.') 
plt.scatter(x1, y1, color='red', label='validation', s=s, marker='.')



# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

# Add legend
plt.legend()
# Set axes equal
plt.axis('equal')

# Show the plot
plt.savefig('scatter_plot.png', dpi=300)
plt.show()

```
